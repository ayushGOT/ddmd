### First Scenario 

```python
chosen_embedded_value = [1.15241539478302,-0.4468420147895813,0.6636378169059753,-0.5994530916213989,0.1852247267961502,-0.1050015464425087]  # Example chosen value
```

```python
# Define a threshold for overlap (you can adjust this threshold)
threshold = 0.5
```

```python
# Calculate the Euclidean distances between the chosen value and all other values
euclidean_distances = []

for embedding in df['embeddings']:
    distance = np.linalg.norm(np.array(embedding) - chosen_embedded_value)
    euclidean_distances.append(distance)

```

```python
# Convert the distances list to a NumPy array
euclidean_distances = np.array(euclidean_distances)
```

```python
# Identify the indices of values that overlap based on the threshold
overlapping_indices = np.where(euclidean_distances < threshold)[0]
```

```python
# Filter the DataFrame to include only overlapping values
overlapping_values_df = df.iloc[overlapping_indices]
```

```python
# Visualize the chosen value and the surrounding values
plt.figure(figsize=(10, 6))
plt.scatter(df['embeddings'].apply(lambda x: x[0]), df['embeddings'].apply(lambda x: x[1]), label='Non-overlapping', alpha=0.5)
plt.scatter(overlapping_values_df['embeddings'].apply(lambda x: x[0]), overlapping_values_df['embeddings'].apply(lambda x: x[1]), color='red', label='Overlapping')
plt.scatter(chosen_embedded_value[0], chosen_embedded_value[1], color='blue', marker='x', s=100, label='Chosen Value')
plt.xlabel('Embedding Dimension 1')
plt.ylabel('Embedding Dimension 2')
plt.title('Visualization of Overlapping and Chosen Embedding Values')
plt.legend()
plt.show()
```

```python
# Print the values that are overlapping with the chosen value
print("Values overlapping with the chosen value:")
print(overlapping_values_df)
```

### Final method using NearestNeighbours

```python
chosen_embedded_value = [1.055411696434021,
  0.08268323540687561,
  -0.26983243227005005,
  -0.8041821718215942,
  -0.657209038734436,
  0.18305988609790802]
```

```python

# Define a threshold for overlap (you can adjust this threshold)
threshold = 0.5
```

```python
# Convert the 'embeddings' column to a NumPy array
embeddings_array = np.array(df['embeddings'].tolist())
```

```python
# Create a NearestNeighbors model
from sklearn.neighbors import NearestNeighbors
# Create a NearestNeighbors model
nn_model = NearestNeighbors(n_neighbors=len(df))
nn_model.fit(embeddings_array)
```

```python
# Find the indices of the nearest neighbors of the chosen embedded value
nearest_indices = nn_model.kneighbors([chosen_embedded_value], n_neighbors=6, return_distance=False)[0]
```

```python

# Extract the overlapping embeddings
overlapping_embeddings = embeddings_array[nearest_indices]
```

```python
# Visualize the chosen value and the overlapping values in the 6-dimensional embedding space
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')  # Specify 3D projection
ax.scatter(embeddings_array[:, 0], embeddings_array[:, 1], embeddings_array[:, 2], label='Non-overlapping', alpha=0.5)
ax.scatter(overlapping_embeddings[:, 0], overlapping_embeddings[:, 1], overlapping_embeddings[:, 2], color='red', label='Overlapping')
ax.scatter(chosen_embedded_value[0], chosen_embedded_value[1], chosen_embedded_value[2], color='blue', marker='x', s=100, label='Chosen Value')
ax.set_xlabel('Embedding Dimension 1')
ax.set_ylabel('Embedding Dimension 2')
ax.set_zlabel('Embedding Dimension 3')
ax.set_title('Visualization of Overlapping Embedding Values in 3D Space')
ax.legend()
plt.show()
```

The above code represnet s nearest neighbors of the chosen embedded value using NearestNeighbors and visualize the chosen value and its nearest neighbors in the original 6-dimensional embedding space using a 3D scatter plot. we can adjust the chosen_embedded_value, n_neighbors, and visualization settings if needed.
