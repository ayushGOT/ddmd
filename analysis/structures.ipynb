```python
mis_df=df[(df.rmsd<4) & (df.rmsd>3) & (df.Q>0.9)]   # misfolded configs
mis_df
```

```python
embed=np.array(mis_df['embeddings'].values.tolist())
embed[3000]
```

```python
from sklearn.cluster import KMeans, MiniBatchKMeans

kmeans = MiniBatchKMeans(
            n_clusters=100, 
            init="k-means++", random_state=42
        ).fit(embed)
```

#### Get the representative points i.e. the points closest to the biggest cluster centroid

```python
# Get cluster labels and centers
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Count the number of data points in each cluster
cluster_sizes = np.bincount(labels)

# Find the label of the largest cluster
largest_cluster_label = np.argmax(cluster_sizes)

# Find the centroid of the largest cluster
largest_cluster_centroid = centroids[largest_cluster_label]

# Find the indices of data points in the largest cluster
largest_points_indices = np.where(labels == largest_cluster_label)[0]

# Get the actual data points in the largest cluster
points = embed[largest_points_indices]

# Calculate the distances from each data point in the largest cluster to the centroid
distances_to_largest_centroid = np.linalg.norm(embed - largest_cluster_centroid, axis=1)

# Find the indices of data points closest to the largest cluster's centroid
closest_points_indices = np.argsort(distances_to_largest_centroid)
print(f"Indices of closest points:{closest_points_indices}")

# Get the actual data points closest to the centroid
closest_points = embed[closest_points_indices]

print(len(closest_points))
print(f"Label of the largest cluster: {largest_cluster_label}")
print(f"Points in the largest cluster:\n{closest_points}")
```

#### Clustering using mdtraj

```python
for index, row in mis_df.iterrows():
    print(row['frame'])
    
import mdtraj as md
traj=[]
for index, row in mis_df.iterrows():
    traj.append(md.load_frame(row['dcd'],row['frame'],top=row['pdb']))
    print(index)
print(traj)
```

```python
# compute all pairwise rmsds between conformations
atom_indices = [a.index for a in traj.topology.atoms if a.element.symbol != 'H']
distances = np.empty((traj.n_frames, traj.n_frames))
for i in range(traj.n_frames):
    distances[i] = md.rmsd(traj, traj, i, atom_indices=atom_indices)

# Transform these distances into similarity scores and determine the centroid
beta = 1
index = np.exp(-beta*distances / distances.std()).sum(axis=1).argmax()
print(index)
centroid = traj[index]
print(centroid)
```
